```python
#saber qual numero é maior

#entrada
a = int(input("digite um numero:"))
b = int(input("digite outro numero:"))

#processamento
if a < b:
 print (f"seu numero {a} é menor que {b}")

if a > b:
 print (f" seu numero {a} é maior que {b}")

if a == b:
 print (f" seus dois numeros {a} e {b} são iguais")



```

```python
#aprovação de aluno nota >= 7 aprovado 5 <= nota < 7 recuperação e o resto é reprovado

#entrada
nota = int(input("digite sua nota:"))

#processamento

if nota >= 7:
 print (f"sua nota foi {nota} vc foi aprovado")

if nota >= 5 and nota < 7:
 print (f"sua nota foi {nota} vc esta de recuperação")

if nota < 5:
 print (f"sua nota foi {nota} vc esta reprovado")




```

```python
#classificação de idade

#entrada
idade = int(input("digite sua idade"))

#processamento

if idade <= 12:
  print (f" você tem {idade} anos, você é criança.")

if idade <= 17 and idade > 12:
 print (f"você tem {idade} anos, é adolescente")

elif idade >= 18 and idade <= 59:
 print (f"vc tem {idade} anos, vc é adulto ")

elif idade > 59:
  print (f"vc tem {idade} vc é idoso")



```

```python
#definir senha fixa

#entrada
senha = int(input("digite uma senha de 4 numeros:"))
senha2 = int(input("confirme sua senha:"))

#processamento
if senha == senha2:
 print (f"sua confirmação esta correta")

if senha != senha2:
 print (f"sua confitmação esta incorreta")



```

```python
#informe se um numero é possitivo negativo ou zero

#entrada
numero = int(input("digite um numero para descobrir se ele é possitivo, negativo ou zero:"))

#processamento
if numero >= 1:
  print (f"seu numero é possitivo")

if numero <= -1:
  print(f"seu numero é negativo")

elif numero == 0:
  print (f"seu numero é zero")
```

```python
#identificar o ano bissexto

#entrada
ano = int(input("digite um ano para descobrir se ele é bissexto:"))

#processamento

ano = int(input('Digite o ano: '))

if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
    print (f'{ano} é um ano bissexto.')
else:
    print (f'{ano} não é um ano bissexto.')
```
