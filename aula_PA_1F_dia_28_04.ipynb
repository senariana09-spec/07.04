```python
#escrver um programa q verifica se a palavra é palindromo e imprimi a palavra

 #entrada
palavra = str(input("digite uma palavra para descobrir se é um palindromo:"))

#processamento
palavra = (palavra)
invertida = ""

for i in (palavra):

  invertida = i + invertida

if palavra == (invertida):
 print (f"{palavra} é um palindromo")

else :
 print (f"{palavra} não é palindromo")


```
