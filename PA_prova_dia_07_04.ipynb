```python
#criar um programa se classifica se uma pessoa é menor de idade, adulto ou idoso

#entrada
idade = int(input("digite sua idade para saber se vc é menor de idade, adulto ou idoso: "))

#processamento

if idade <= 17:
  print (f"você tem {idade} anos, você é meno de idade")

elif idade >= 18 and idade <= 59:
  print (f"você tem {idade} anos, você é adulto")

else:
  print(f"você tem {idade} anos, você é idoso")





```

```python
#programa que le o valor da compra e aplica o desconto conforme as regras
#ate 100 reais sem desconto de 101 a 500 10% de desconto e acima de 500 reais 20% de desconto

#entrada
compra = int(input("digite o valor da sua compra para descobrir seu desconto"))


#processamento
if compra <= 100:
  print (f"você não terá desconto:")

elif compra >= 101 and compra <= 500:

  print (f"sua compra adquiriu 10% de desconto:")

else:
  print (f"sua compra adquiriu 20% de desconto:")





```

```python
# classificar energia do personagem

#entrada
energia = int(input("digite a energia do seu personagem:"))

#processamento
if energia == 0:
  print (f"seu personagem tem energia igual a {energia} ele está morto")

if energia >= 1 and energia <= 30:
  print (f"seu personagem tem energia igual a {energia} ele está critico")

elif energia >= 31 and energia <= 70:
  print (f"seu personagem tem a energia igual a {energia} normal")

elif energia >= 71 and energia <= 100:
  print(f" seu personagem tem a energia igual a {energia} ele está saldavel")
```
